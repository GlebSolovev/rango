```python
from evaluation.eval_analysis import find_shared_proofs, get_times_and_num_proofs, get_expansions_and_num_proofs
import matplotlib.pyplot as plt
```

```python
EVAL_DIRS = [
    ("Base", "/home/ubuntu/coq-modeling/evals/eval/codellama-7b-base"),
    ("Base with Premises", "/home/ubuntu/coq-modeling/evals/eval/codellama-7b-base-premise"),
    ("Fine-Tuned", "/home/ubuntu/coq-modeling/evals/eval/codellama-7b-basic"),
    ("Fine-Tuned with Premises", "/home/ubuntu/coq-modeling/evals/eval/codellama-7b-premise"),
]
```

```python
eval_dirs = [b for a, b in EVAL_DIRS]
shared_proof_names = find_shared_proofs(eval_dirs)
```

```python
fig, ax = plt.subplots()
assert type(ax) == plt.Axes
for eval_name, eval_path in EVAL_DIRS:
    times, num_proofs = get_times_and_num_proofs(eval_path, shared_proof_names)
    ax.plot(times, num_proofs, label=eval_name)
ax.legend()
ax.set_xlabel("Time(s)")
ax.set_ylabel("# Proofs Found")
ax.set_title(f"# Proofs found out of {len(shared_proof_names)} Theorems.")

```

```python
fig, ax = plt.subplots()
assert type(ax) == plt.Axes
for eval_name, eval_path in EVAL_DIRS:
    times, num_proofs = get_expansions_and_num_proofs(eval_path, shared_proof_names)
    ax.plot(times, num_proofs, label=eval_name)
ax.legend()
ax.set_xlabel("# Expansions")
ax.set_ylabel("# Proofs Found")
ax.set_title(f"# Proofs Found out of {len(shared_proof_names)} Theorems.")
```
