```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

labels = ["0-0.1", "0.1-0.2", "0.2-0.3", "0.3-0.4", "0.4-0.5", "0.5-0.6", "0.6-0.7", "0.7-0.8", "0.8-0.9", "0.9-1"]
zero_to_ten_counts = [1865, 46, 4017, 13139, 2199, 21455, 30805, 33415, 48107, 25212]
ten_to_twenty_counts = [0, 3, 14, 29, 71, 316, 1680, 6160, 31988, 9969]
```

```python
fig, ax = plt.subplots()
ticks = np.arange(len(labels))
width = 0.4
ax.set_xticks(ticks)
ax.bar(ticks - (width / 2), zero_to_ten_counts, width=width, label="0-10 Step Proofs")
ax.bar(ticks + (width / 2), ten_to_twenty_counts, width=width, label="10-20 Step Proofs")
ax.set_xticklabels(labels, rotation=30, ha="right")
ax.set_ylabel("# Pairs of Proofs")
ax.set_xlabel("Normalized Levenshtein Distance")
ax.set_title("Frequency of Similar Proofs")
ax.legend()
```
